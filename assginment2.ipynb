# data replacement

```python
data['MultipleLines']   .replace({'No':'No','Yes':'Yes','No phone service':'No'},inplace=True)
data['OnlineSecurity']  .replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
data['OnlineBackup']    .replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
data['DeviceProtection'].replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
data['TechSupport']     .replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
data['StreamingTV']     .replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
data['StreamingMovies'] .replace({'No':'No','Yes':'Yes','No internet service':'No'},inplace=True)
```

# Missing values treatment



```python
data.isna().sum()
```

```python
data.dropna(subset=["SeniorCitizen"],inplace=True)
t_mean = data['tenure'].mean()
fills  = {"tenure":t_mean}
data.fillna(fills,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop("customerID",axis=1,inplace=True) #drop index column
```

# label Encoder

```python
from sklearn.preprocessing import LabelEncoder #importing labelEncoder
```

```python
d={}                                                # new data dictionary for saveing labelEncoders objects
for i in data.columns :                             # looping data2 coluoms  
    if data[i].dtype == "object":                   # select only object attributes 
        d[i]=LabelEncoder()                         #new labelencoder object for each attribute with object datatype
        d[i].fit(data[i])                           #fitting dataframe attribute
        data[i]=d[i].transform(data[i])             #transform attribute and save it in dataframe
```

# Standardization
## z-score

```python
data["tenure"]        =(data["tenure"]-data["tenure"].mean())/data["tenure"].std() #standarization z-score=(X-mean)/std
data["MonthlyCharges"]=(data["MonthlyCharges"]-data["MonthlyCharges"].mean())/data["MonthlyCharges"].std()
data["TotalCharges"]  =(data["TotalCharges"]-data["TotalCharges"].mean())/data["TotalCharges"].std()
```

# KNN

```python
X = data.iloc[:, 2:9].values
y = data.iloc[:, 10].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnclassifier = KNeighborsClassifier(n_neighbors = 15, metric = 'minkowski', p = 2)
knnclassifier.fit(X_train, y_train)
```

```python
print("Test set predictions:", knnclassifier.predict(X_test))
```

```python
print("Test set accuracy: {:.2f}".format(knnclassifier.score(X_test, y_test)))
```

```python
y_pred = knnclassifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
